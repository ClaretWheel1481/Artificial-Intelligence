```python
import pandas as pd
s = pd.Series([1,3,5,9,6,0.8])
print(s)
```

```python

print("列表创建Series")
s1 = pd.Series([1,2,1,8,1])
print(s1)
print("字典创建Series")
s2 = pd.Series({'Longitude':41,'Latitude':118,'Temperature':23})
print("First Value in s2:",s2['Longitude'])
print("用序列做Series索引")
s3 = pd.Series([3.4,0.8,2.1,0.3,1.5],range(5,10))
print('First Value in s3:',s3[5])
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax1 = fig.add_subplot(3,2,1)
ax2 = fig.add_subplot(4,2,2)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
rect = plt.Rectangle((0.2,0.75),0.4,0.15,color = 'red',alpha = 0.7)
circ = plt.Circle((0.7,0.2),0.15,color='blue',alpha = 0.3)
pgon = plt.Polygon([[0.15,0.15],[0.35,0.4],[0.2,0.6]],color = 'green',alpha = 0.9)
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
a = np.arange(10)
plt.xlabel('x')
plt.ylabel('y')
plt.plot(a,a*1.5,a,a*2.5,a,a*3.5,a,a*4.5)
plt.legend(['1.5x','2.5x','3.5x','4.5x'])
plt.title("Simple Lines")
plt.show()
```
