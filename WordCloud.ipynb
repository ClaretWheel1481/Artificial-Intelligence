```python
import re
import collections
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from wordcloud import WordCloud
```

```python
# Dataset
string_data = """
Here is your text content, which can be anything, such as a novel or an article. This text is just an example,
and should be replaced with the text you actually want to analyze.
"""

# Words Preprocessing
pattern = re.compile(u'\t|\n|\.|-|:|;|\)|\(|\?|')
string_data = re.sub(pattern,'',string_data)

# Space subscript processing
words = string_data.split()

# Words that should be removed
remove_words = ['the','and','is','with','for','as','on','of','at','to','by','a','an','in','that','this','it','these','those','such']

# Removing words from "remove_words"
filtered_words = [word for word in words if word.lower() not in remove_words]

# Words frequency statistics
word_counts = collections.Counter(filtered_words)

# White Background
mask = np.array(Image.new("RGB",(800,800),(255,255,255)))

# WordCloud Settings
wc = WordCloud(
    background_color = 'white', # Background color
    max_words = 100,  # The max words
    max_font_size = 100,  # The max fontsize
)

# Generated word cloud
wc.generate_from_frequencies(word_counts)
plt.imshow(wc,interpolation = 'bilinear')
plt.axis('off')
plt.show()
```
